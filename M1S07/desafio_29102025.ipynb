```python
import pandas as pd
```

```python
df_clamed = pd.read_csv(r'C:\Users\marco\Downloads\dataset_clamed.csv')
```

```python
# Remover duplicadas
df_clamed.drop_duplicates(inplace=True)

# Preencher valores ausentes em 'preco', 'quantidade_estoque' e 'fornecedor'
df_clamed.fillna({'preco': df_clamed['preco'].mean().round(2), 'quantidade_estoque': 1, 'fornecedor' : 'Não informado'}, inplace=True)
```

```python
# Função para remover outliers usando o método de Tukey
def remover_outliers_tukey(df, coluna):
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    return df[
        (df[coluna] >= limite_inferior) &
        (df[coluna] <= limite_superior)
    ]

# Usando a função nas colunas 'preco' e 'quantidade_estoque'
df_clamed = remover_outliers_tukey(df_clamed, 'preco')
df_clamed = remover_outliers_tukey(df_clamed, 'quantidade_estoque')
```

```python
# Fazendo a normalização dos dados usando o Min-Max

preco_min = df_clamed['preco'].min()
preco_max = df_clamed['preco'].max()
df_clamed['preco_norm'] = (df_clamed['preco'] - preco_min) / (preco_max - preco_min)
```
